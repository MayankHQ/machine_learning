```python
# Raw Data
ages = [22, 25, 21, 23, 24, 22, 500, 23, 21]
def mean(ages):
  return sum(ages)/len(ages)

def median(ages):
  sorted(ages)
  return ages[len(ages)//2]

print(mean(ages))
print(median(ages))
```

```python
from scipy import stats

group_a = [5, 6, 5, 7, 6] # Sugar pill (Slow recovery)
group_b = [2, 3, 2, 4, 3] # Medicine (Fast recovery)

# This runs a T-Test to compare the two groups
t_stat, p_value = stats.ttest_ind(group_a, group_b)

print("P-value:", p_value)
```

```python
list1 = [2,5,7]
list2 = [0.4, 8,4]
def dot_product(list1, list2):
  total = 0
  for i in range(len(list1)):
    total += list1[i] * list2[i]
  return total
print(dot_product(list1, list2))
```

```python
prices = ["$10", "$20", "$50", "$5"]

# The Logic: "For every 'p' in 'prices', replace '$' with nothing, then turn to int"
clean_prices = [int(p.replace("$", "")) for p in prices]

print(clean_prices)
# Output: [10, 20, 50, 5]
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset
np.random.seed(42)
n_movies = 200

data = {
    "Title": [f"Movie_{i}" for i in range(n_movies)],
    "Genre": np.random.choice(["Action", "Comedy", "Drama", "Horror"], n_movies),
    "Budget_Million": np.random.randint(10, 300, n_movies), # Cost to make
    "Revenue_Million": np.random.randint(5, 800, n_movies), # Money made
    "IMDB_Rating": np.round(np.random.uniform(3.0, 9.5, n_movies), 1),
    "Year": np.random.randint(2000, 2025, n_movies)
}

df = pd.DataFrame(data)

# INTRODUCING DIRTY DATA (You must fix these!)
# 1. Some movies have 0 revenue (Data error)
df.loc[0:4, "Revenue_Million"] = 0
# 2. Some movies have no Genre (Missing values)
df.loc[10:14, "Genre"] = np.nan

print("Dataset Created. Rows:", len(df))
print(df.head())
```

```python
remove_df = df.replace(0, np.nan).dropna() #task1
```

```python
remove_df["Profit"] = remove_df["Revenue_Million"] - remove_df["Budget_Million"]  #task2
```

```python
sns.boxplot(data = remove_df, x = "Genre", y= "Profit") #task3
plt.show()
```

```python
trend_data = remove_df.groupby("Year")["Budget_Million"].mean()
plt.plot(trend_data.index, trend_data.values, marker = 'o')
plt.xlabel("Year")
plt.ylabel("Budget")
plt.show()
```

```python
corr_m = remove_df.corr(numeric_only=True)
sns.heatmap(corr_m, annot=True, cmap="coolwarm")
plt.show()
```
